## Descripción de los datos
### Los Juegos Olímpicos

Los juegos olímpicos (JJ.OO.) son el mayor evento deportivo internacional y multidisciplinario en el que participan atletas de varios países del mundo. Existen dos ediciones de los JJ.OO., los de Verano y los de invierno, que se celebran con un intervalo de dos años entre sí. Esto evento fueron inspirados en los Juegos Olímpicos de la antigüedad del Siglo VIII a. e. c., organizados por la antigua Grecia.

La primera edición de los Juegos Olímpicos de la era moderna se llevó a cabo en Atenas, capital de Grecia, a partir del 6 de abril de 1896. Desde aquella oportunidad, han sido realizados cada cuatro años en diversas ciudades del mundo, siendo las únicas excepciones las ediciones de 1916, 1940 y 1944, debido al estallido de la Primera y Segunda Guerra Mundial; así como la postergación de la de 2020 para 2021, debido a la pandemia de COVID-19. Fuente: [Wikipedia en español](https://es.wikipedia.org/w/index.php?title=Juegos_Ol%C3%ADmpicos&oldid=162341811).

### Distribución

Comencemos con el análisis de la distribución, para lo cual recurriremos al histograma. Para cada edición de los JJ.OO., haremos un histograma para cada metal, oro, plata y bronce. Para culminar, haremos lo propio para las medallas obtenidas por ambas ediciones en conjunto. Para crear los histogramas, la siguiente función será de ayuda,

```python
def crear_histograma(data, campos, titulo="", figsize=(8,12), bins="auto"):
	if not isinstance(campos, (list, tuple)):
		campos = [campos]

	nsubs = len(campos)

	fig, ax = plt.subplots(nrows=nsubs, figsize=figsize)

	if nsubs == 1:
		ax = [ax]

	for index,campo in enumerate(campos):
		sns.histplot(data=data, x=campo, bins=bins, ax=ax[index])

	fig.subplots_adjust(wspace=0.5)
	fig.suptitle(titulo, y=0.93)

	plt.show()
```

#### JJ.OO. de Verano

Los histogramas para las medallas de los JJ.OO. de Veranos son,

```python
crear_histograma(
				  df_verano,
				  campos=["gold_summer", "silver_summer", "bronze_summer"],
				  titulo="Histograma. Total medallas ganadas por país en los JJ.OO. de Verano",
				  figsize=(8,15)
				)
```

**Análisis**

* Valores atípicos muy extremos. Aunque los tres histogramas aún son visibles, es bastante notorio la existencia de un bin con valores bastante extremos.
* La distribución de los tres metales es muy similar: Una distribución sesgada a la derecha.
* Presenta un pico en el primer bin, para luego descender bruscamente.
* Existe un cúmulo de datos concentrados a la derecha del histograma para los tres metales; el cúmulo comienza en 0 y termina alrededor de las 100 medallas obtenidas. Luego de eso la distribución pierde densidad.

En los histogramas se puede observar como existe una gran desigualdad en las medallas obtenidas. Son pocos los equipos los que monopolizan los tres primeros lugares de las competencias, dejando a la mayoría con un desempeño sutil.

## Análisis de Correlación

En esta sección, se calcularán matrices de correlación que se representarán mediante gráficos de calor para facilitar su interpretación visual. Inicialmente, se analizará la correlación entre las medallas obtenidas —oro, plata y bronce— en las ediciones de los Juegos Olímpicos de Verano e Invierno por separado, con el fin de identificar posibles patrones de rendimiento. Posteriormente, se integrarán ambas ediciones para examinar si existe una correlación significativa entre el desempeño de los países en los Juegos de Verano y los de Invierno.

Además, se incorporarán indicadores externos con el objetivo de explorar posibles relaciones entre el desempeño olímpico y el nivel socioeconómico y político de cada país. Entre estos indicadores se incluyen variables como el PIB, la población, el gasto público en educación, entre otros factores clave que podrían influir en los resultados obtenidos en los Juegos Olímpicos.

Para la construcción de las gráficas de calor se usará la siguiente función,

```python
def matriz_correlacion(corr_matrix, figsize=(10, 8), title="Matriz de correlación", annot=True):
	fig, ax = plt.subplots(figsize=figsize)

	sns.heatmap(corr_matrix, cmap='RdBu_r', center=0, vmin=-1, vmax=1, annot=annot, ax=ax)
	ax.set_title(title)

	plt.show()
```

La matriz generada mostrará el nivel de correlación usando una escala de colores de azul, para las correlaciones negativas, a rojo, para las correlaciones positivas, y blanco para correlación nula:

```python
sns.color_palette("RdBu_r", as_cmap=True)
```

### Matriz de correlación JJ.OO. de Verano

Comencemos con el análisis de correlación de los metales obtenidos en los JJ.OO. de Verano,

```python
columns_summer = ["Num_Summer", "gold_summer", "silver_summer", "bronze_summer"]

matriz_correlacion(df_verano[columns_summer].corr(), title="Matriz de Correlación JJ.OO. de Verano")
```

**Análisis**

 - Existe un Correlación Perfecta entre los tres metales, lo que indica que los países que tienen buen desempeño, lo tienen en las tres medallas.
 - Existe una Correlación Débil entre el número de ediciones en las que se participaron y el desempeño olímpico. Esto significa que el número de ediciones participadas tiene poco efecto en el desempeño olímpico.

### Matriz de correlación entre ambas ediciones

Ahora queremos explorar las correlaciones entre ambas ediciones, por lo que crearemos una matriz de correlación aumentada. En este caso usaremos el dataframe original `df_medallas`, para incluir los registros que dejamos afuera para el estudio de las ediciones por separado.

Correlación de Débil a Moderada ($0.45$ - $0.57$). Esto podría indicar que muchos países con un buen desempeño en la edición de Verano, suelen enviar representación a la edición de Invierno.
 - La correlación entre el desempeño al obtener los tres metales en ambas ediciones (`gold_summer`, `silver_summer`, `bronce_summer`, `gold_winter`, `silver_winter`, `bronce_winter`) entre ambos países es Moderada ($0.61$ - $0.67$).
