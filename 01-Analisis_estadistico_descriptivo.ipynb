## Introducción
### Descripción del proyecto

Este proyecto tiene como objetivo analizar el rendimiento de cada país en los Juegos Olímpicos (JJ.OO.), tanto en su edición de Verano como de Invierno. Se ha recopilado datos sobre el total de medallas obtenidas por cada equipo olímpico desde la primer edición, así como una variedad de factores geográficos, como el continente y la región a la que pertenece el país, y factores socioeconómicos, como la población y el PIB. ### Objetivos del proyecto

Los objetivos del proyecto son los siguientes:

 - Identificar patrones de éxito olímpico en los diferentes continentes y regiones del mundo.

## Descripción de los datos
### Los Juegos Olímpicos

Los juegos olímpicos (JJ.OO.) son el mayor evento deportivo internacional y multidisciplinario en el que participan atletas de varios países del mundo. Existen dos ediciones de los JJ.OO., los de Verano y los de invierno, que se celebran con un intervalo de dos años entre sí. Esto evento fueron inspirados en los Juegos Olímpicos de la antigüedad del Siglo VIII a. e. c., organizados por la antigua Grecia.

La primera edición de los Juegos Olímpicos de la era moderna se llevó a cabo en Atenas, capital de Grecia, a partir del 6 de abril de 1896. Desde aquella oportunidad, han sido realizados cada cuatro años en diversas ciudades del mundo, siendo las únicas excepciones las ediciones de 1916, 1940 y 1944, debido al estallido de la Primera y Segunda Guerra Mundial; así como la postergación de la de 2020 para 2021, debido a la pandemia de COVID-19. Fuente: [Wikipedia en español](https://es.wikipedia.org/w/index.php?title=Juegos_Ol%C3%ADmpicos&oldid=162341811).

Por otra parte, tenemos el dataframe guardado en `info_countries.csv`. Este dataframe cuenta con información obtenido desde Wikipedia, y del *World Bank Group*. Podemos clasificar los campos de este dataframe en dos categorías: Parámetros e Indicadores. Los Parámetros son aquellos que resumen información como el nombres, códigos de identificación y otros datos más de clasificación. Los campos del dataframe que son parámetros son los siguientes,
 - `nombre`: (`str`) Nombre en español del país.

 - `WB-2_code`: (`str`) Código de dos letras que se usa para identificar países en los sistemas del World Bank Group.

Luego, tenemos los indicadores. Estos son métricas estadísticas alguna característica del país. Estos valores, generalmente numéricos, son extraídos del *World Bank Group*, que a su vez obtiene la información de las dependencias de información de los países. Los campos que pertenecen a esta clasificación son,
 - `Access_to_electricity_media`: (`float`) Media histórica del indicador desde 1990 hasta 2022:
   - **Nombre del indicador en inglés**: *Access to electricity (% of population)*
   - **Código del indicador**: EG.ELC.ACCS.ZS
   - **Descripción**: Porcentaje de la población con acceso a la electricidad. Los datos sobre se recopilan de la industria, encuestas nacionales y fuentes internacionales.
 - `AFFVA_media`: (`float`) Media histórica del indicador desde 1960 hasta 2023:
   - **Nombre del indicador en inglés**: *Agriculture, forestry, and fishing, value added (% of GDP)*.
   - **Código del indicador**: NV.AGR.TOTL.ZS
   - **Descripción**: Mide la contribución porcentual del sector agrícola, silvicultural y pesquero al Producto Interno Bruto (PIB o GDP por sus siglas en inglés) en porcentaje.
 - `Control_of_Corruption_media`: (`float`) Media histórica del indicador desde 1996 hasta 2022:
   - **Nombre del indicador en inglés**: *Control of Corruption: Estimate*.
   - **Código del indicador**: CC.EST
   - **Descripción**: Captura percepciones de hasta qué punto el poder público se ejerce para beneficio privado, incluidas formas tanto pequeñas como grandes de corrupción, así como la "captura" del Estado por parte de élites e intereses privados. La estimación proporciona la puntuación del país en el indicador agregado, en unidades de una distribución normal estándar, es decir, entre aproximadamente $-2.5$ y $2.5$.
 - `Expenditure_on_education_media`: (`float`) Media histórica del indicador desde 1970 hasta 2023:
   - **Nombre del indicador en inglés**: *Government expenditure on education, total (% of GDP)*.
   - **Código del indicador**: SE.XPD.TOTL.GD.ZS
   - **Descripción**: El gasto del gobierno general en educación (corriente, capital y transferencias) se expresa como porcentaje del PIB. Incluye gastos financiados por transferencias de fuentes internacionales al gobierno. El gobierno general generalmente se refiere a los gobiernos locales, regionales y centrales.
 - `GDP_per_capita_media`: (`float`) Media histórica del indicador desde 1960 hasta 2023:
   - **Nombre del indicador en inglés**: *GDP per capita (current US$)*.
   - **Código del indicador**: NY.GDP.PCAP.CD
   - **Descripción**: El PIB per cápita es el producto interno bruto dividido por la población a mitad de año. El PIB es la suma del valor agregado bruto de todos los productores residentes en la economía más los impuestos sobre los productos y menos los subsidios no incluidos en el valor de los productos. Se calcula sin hacer deducciones por depreciación de activos fabricados o por agotamiento y degradación de recursos naturales. Los datos están en dólares estadounidenses actuales.
 - `GDP_per_capitaPPP_media`: (`float`) Media histórica del indicador desde 1990 hasta 2023:
   - **Nombre del indicador en inglés**: *GDP per capita, PPP (current international $)*.
   - **Código del indicador**: NY.GDP.PCAP.PP.CD
   - **Descripción**: Este indicador proporciona valores per cápita del producto interno bruto (PIB) expresado en dólares internacionales corrientes convertidos mediante el factor de conversión de paridad de poder adquisitivo (PPA). Es más adecuado para comparar el poder adquisitivo y el nivel de vida entre países, ajustando las diferencias en los costos de vida y los precios.
 - `population_media`: (`float`) Media histórica del indicador desde 1960 hasta 2023:
   - **Nombre del indicador en inglés**: *Population, total*.
   - **Código del indicador**: SP.POP.TOTL
   - **Descripción**: La población total se basa en la definición de facto de población, que cuenta a todos los residentes independientemente de su estatus legal o ciudadanía. Los valores mostrados son estimaciones de mitad de año.
 - `surface_area_media`: (`float`) Media histórica del indicador desde 1961 hasta 2021:
   - **Nombre del indicador en inglés**: *Surface area (sq. km)*.
   - **Código del indicador**: AG.SRF.TOTL.K2
   - **Descripción**: Por superficie se entiende la superficie total de un país, incluidas las zonas cubiertas por masas de agua interiores y algunas vías navegables costeras. Se mide en $km^2$ (kilómetros cuadrados).

### Distribución

Comencemos con el análisis de la distribución, para lo cual recurriremos al histograma. Para cada edición de los JJ.OO., haremos un histograma para cada metal, oro, plata y bronce. Para culminar, haremos lo propio para las medallas obtenidas por ambas ediciones en conjunto. Para crear los histogramas, la siguiente función será de ayuda,

```python
def crear_histograma(data, campos, titulo="", figsize=(8,12), bins="auto"):
	if not isinstance(campos, (list, tuple)):
		campos = [campos]

	nsubs = len(campos)

	fig, ax = plt.subplots(nrows=nsubs, figsize=figsize)

	if nsubs == 1:
		ax = [ax]

	for index,campo in enumerate(campos):
		sns.histplot(data=data, x=campo, bins=bins, ax=ax[index])

	fig.subplots_adjust(wspace=0.5)
	fig.suptitle(titulo, y=0.93)

	plt.show()
```

## Análisis de Correlación

En esta sección, se calcularán matrices de correlación que se representarán mediante gráficos de calor para facilitar su interpretación visual. Inicialmente, se analizará la correlación entre las medallas obtenidas —oro, plata y bronce— en las ediciones de los Juegos Olímpicos de Verano e Invierno por separado, con el fin de identificar posibles patrones de rendimiento. Posteriormente, se integrarán ambas ediciones para examinar si existe una correlación significativa entre el desempeño de los países en los Juegos de Verano y los de Invierno.

Además, se incorporarán indicadores externos con el objetivo de explorar posibles relaciones entre el desempeño olímpico y el nivel socioeconómico y político de cada país. Entre estos indicadores se incluyen variables como el PIB, la población, el gasto público en educación, entre otros factores clave que podrían influir en los resultados obtenidos en los Juegos Olímpicos.

Para la construcción de las gráficas de calor se usará la siguiente función,

```python
def matriz_correlacion(corr_matrix, figsize=(10, 8), title="Matriz de correlación", annot=True):
	fig, ax = plt.subplots(figsize=figsize)

	sns.heatmap(corr_matrix, cmap='RdBu_r', center=0, vmin=-1, vmax=1, annot=annot, ax=ax)
	ax.set_title(title)

	plt.show()
```

La matriz generada mostrará el nivel de correlación usando una escala de colores de azul, para las correlaciones negativas, a rojo, para las correlaciones positivas, y blanco para correlación nula:

```python
sns.color_palette("RdBu_r", as_cmap=True)
```

### Matriz de correlación JJ.OO. de Verano

Comencemos con el análisis de correlación de los metales obtenidos en los JJ.OO. de Verano,

```python
columns_summer = ["Num_Summer", "gold_summer", "silver_summer", "bronze_summer"]

matriz_correlacion(df_verano[columns_summer].corr(), title="Matriz de Correlación JJ.OO. de Verano")
```

**Análisis**

 - Existe un Correlación Perfecta entre los tres metales, lo que indica que los países que tienen buen desempeño, lo tienen en las tres medallas.
 - Existe una Correlación Débil entre el número de ediciones en las que se participaron y el desempeño olímpico. Esto significa que el número de ediciones participadas tiene poco efecto en el desempeño olímpico.

### Matriz de correlación JJ.OO. de Invierno

Continuemos con la Matriz de correlación de los Juegos Olímpicos de Invierno,

```python
columns_winter = ["Num_Winter", "gold_winter", "silver_winter", "bronze_winter"]

matriz_correlacion(df_invierno[columns_winter].corr(), title="Matriz de Correlación JJ.OO. de Invierno")
```

**Análisis**

 - De nueva cuenta, tenemos una Correlación Perfecta entre el desempeño olímpico para cada metal, lo que refuerza que los países obtienen aproximadamente la misma cantidad de medallas de los tres metales.
